```python
# Create a sample dataset
import numpy as np
import pandas as pd
data = {
    "Name": ["Janu", "Prassu", np.nan, "Indu"],
    "Age": [19, np.nan, 19, 19],
    "Score":[85, 90, np.nan, 88],
}
df = pd.DataFrame(data)
print("Original DataFrame: \n", df)
# Filling missing values using fillna and interpolate
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Name"] = df["Name"].fillna("Unknown")
df["Score"] = df["Score"].interpolate()
print("Dataset: \n", df)
# Renaming columns
df = df.rename(columns={"Name":"Student_name", "Score":"Exam_score"})
print("Dataset: \n", df)
```

```python
# Exercise1:Clean a dataset by handling missing values and renaming columns
import numpy as np
import pandas as pd
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Janu", "Prassu", "Indu"],
    "Age": [19, 19, 19],
})
df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [85, 90, 88],
})
print("Dataset 1:\n", df1)
print("Dataset 2:\n", df2)
merged = pd.merge(df1, df2, on="ID")
print("Merged dataset:\n", merged)
```
