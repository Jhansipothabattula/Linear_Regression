# Getting Started with PyTorch



**Creating and Running Your First PyTorch Program**

```python
import torch

# Create two random tensors
tensor_a = torch.rand(3, 3)
tensor_b = torch.rand(3, 3)

# Perform matrix multiplication
result = torch.matmul(tensor_a, tensor_b)

# Print the tensors and the result
print("Tensor A:")
print(tensor_a)
print("\nTensor B:")
print(tensor_b)
print("\nResult of matrix multiplication:")
print(result)

# Exploring Further: Moving to GPU

if torch.cuda.is_available():
    tensor_a = tensor_a.to('cuda')
    tensor_b = tensor_b.to('cuda')
    result = torch.matmul(tensor_a, tensor_b)
    print("\nResult on GPU:")
    print(result)

```
