# Overview of TensorFlow

### 1. What is TensorFlow?

TensorFlow is an open-source library developed by the **Google Brain** team for numerical computation and large-scale machine learning. It uses a system of **Tensors** (multi-dimensional arrays) that flow through a **Graph** of operations, allowing you to build and train neural networks efficiently.

### 2. Key Features

* **Flexibility:** Allows you to build models using high-level abstractions or dive deep into custom mathematical operations.
* **Scalability:** You can run your code on a single CPU, multiple GPUs, or even Googleâ€™s custom **TPUs** (Tensor Processing Units).
* **High-Level APIs (Keras):** Makes building models as simple as stacking layers (e.g., `layers.Dense`).
* **TensorBoard:** A powerful visualization tool used to track metrics like loss and accuracy, or to visualize the model graph.

### 3. TensorFlow Components

TensorFlow is more than just a library; it is a full ecosystem for the entire machine learning lifecycle:

| Component | Purpose |
| --- | --- |
| **TF Core** | The foundational engine for building and executing ML models. |
| **TF Extended (TFX)** | An end-to-end platform for deploying production ML pipelines (data validation, training, serving). |
| **TF Lite** | A lightweight solution for deploying models on **mobile devices** and IoT/embedded systems. |
| **TensorFlow.js** | A library for training and deploying models in the **browser** or on Node.js. |
