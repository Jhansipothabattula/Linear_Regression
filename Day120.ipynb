# Loss Functions and Optimizers


## Loss Functions and Optimizers

* ### Loss Functions


* #### Categorical Cross-Entropy

$$-\frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{m} y_{ij} \log(\hat{y}_{ij})$$


`tf.keras.losses.categorical_crossentropy()`
* #### Sparse Categorical Cross-Entropy

$$-\frac{1}{n} \sum_{i=1}^{n} \log(\hat{y}_{i}[y_{i}])$$


`tf.keras.losses.sparse_categorical_crossentropy()`







## **Loss Functions and Optimizers**

* ### **Optimizers**


* **Stochastic Gradient Descent (SGD)**
* TensorFlow Usage: `tf.keras.optimizers.SGD(learning_rate=0.01)`


* **Adam**
* TensorFlow Usage: `tf.keras.optimizers.Adam(learning_rate=0.001)`


* **RMSProp**
* TensorFlow Usage: `tf.keras.optimizers.RMSprop(learning_rate=0.001)`


* **Adagrad**
* TensorFlow Usage: `tf.keras.optimizers.Adagrad(learning_rate=0.01)`



### Custom Loss and Optimizer Implementation

```python
import tensorflow as tf

class CustomLoss(tf.keras.losses.Loss):
    def __init__(self):
        super(CustomLoss, self).__init__()

    def call(self, y_true, y_pred):
        # Implement custom loss calculation
        loss_value = tf.reduce_mean(tf.square(y_true - y_pred))
        return loss_value

custom_loss = CustomLoss()

custom_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
