#Data visualization with Matplotlib and Seaborn


**Matplotlib is a Python library used to create graphs and visualizations**

```python
# Line plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.show()
```

**1. A line plot is a graph that displays data points connected by straight lines to show trends over time or sequence**

```python
# Line plot
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [10, 20, 30], label = "Trend")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

**2. A bar chart is a graph that uses rectangular bars to compare different categories or values**

```python
# Bar chart
import matplotlib.pyplot as plt
Categories = ["A", "B", "C"]
Values = [10, 15, 7]
plt.bar(Categories, Values, color="purple")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

**3. A histogram is a graph that shows the distribution of numerical data by grouping values into bins**

```python
# Histogram
import matplotlib.pyplot as plt
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color="pink", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

**4. A scatter plot is a graph that shows the relationship between two variables using dots on a coordinate plane**

```python
# Scatter Plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 12, 25, 30, 45]
plt.scatter(x, y, color="red")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Dataset 1"])
plt.show()
```

**A seaborn plot is a statistical visualization created using the Seaborn library, which makes charts more attractive and easier to interpret**


**1. A heatmap is a chart that uses colors to represent values in a matrix or table.**


```python
# Heatmap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = np.random.rand(5, 5)
sns.heatmap(data, annot = True, cmap = "coolwarm")
plt.title("Heatmap")
plt.show()
```

**Exercise 1 :- Create basic plots with Matplotlib**

```python
# Line Plot
import matplotlib.pyplot as plt
years = [2010, 2011, 2012, 2013]
sales = [100, 120, 140, 160]
plt.plot(years, sales, label="Sales Trend", color="blue", marker="x")
plt.title("Sales over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# Bar Chart
categories = ["Electronics", "Clothing", "Groceries"]
revenue = [250, 400, 150]
plt.bar(categories, revenue, color="violet")
plt.title("Revenue by Category")
plt.xlabel("Categories")
plt.ylabel("Revenue")
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
hours_studied = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 85]
plt.scatter(hours_studied, exam_scores, color="orange")
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Scores")
plt.show()
```
