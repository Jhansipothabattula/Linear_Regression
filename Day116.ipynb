# Creating and Running a TensorFLow Session

- TensorFLow 2.x(Eager Execution)

```python
import tensorflow as tf

# Define TensorFlow operations
a = tf.constant(2)
b = tf.constant(3)
c = tf.add(a, b)

with tf.compat.v1.Session() as sess:
    result = sess.run(c)
print("Result: \n", result)
```
